Task 7: Please make bins based on age group, then print mean of Adspend, Conversion Rate and Website visits.

```python
def assign_age_group(age):
  if age <= 25:
    return 'Group 1 (0-25)'
  elif 26 <= age <= 35:
    return 'Group 2 (26-35)'
  elif 36 <= age <= 45:
    return 'Group 3 (36-45)'
  elif 46 <= age <= 55:
    return 'Group 4 (46-55)'
  elif 56 <= age <= 65:
    return 'Group 5 (56-65)'
  else:
    return 'Group 6 (66+)'

df['AgeGroup'] = df['Age'].apply(assign_age_group)
grouped_age_summary = df.groupby('AgeGroup')[['AdSpend', 'ConversionRate', 'WebsiteVisits']].mean()
print(grouped_age_summary)
```
